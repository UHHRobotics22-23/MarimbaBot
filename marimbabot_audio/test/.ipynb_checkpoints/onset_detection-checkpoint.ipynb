```python
import librosa
import rosbag
import os
import rospy
import numpy as np
import struct
import pretty_midi
import librosa.display
import matplotlib.pyplot as plt
```

```python
plt.style.use('seaborn-muted')
plt.rcParams['figure.figsize'] = (14, 5)
plt.rcParams['axes.grid'] = True
plt.rcParams['axes.spines.left'] = False
plt.rcParams['axes.spines.right'] = False
plt.rcParams['axes.spines.bottom'] = False
plt.rcParams['axes.spines.top'] = False
plt.rcParams['axes.xmargin'] = 0
plt.rcParams['axes.ymargin'] = 0
plt.rcParams['image.cmap'] = 'gray'
# plt.rcParams['image.interpolation'] = None
```

```python
root = "/home/wang/workspace/marimbabot_ws/audio_patch2/src/marimbabot/marimbabot_audio/test"
bag_file = os.path.join(root,"music.bag")
sr=44100
hop_length=512
fmin = pretty_midi.note_number_to_hz(pretty_midi.note_name_to_number("C4"))
semitones = 36+24
```

```python
rospy.init_node("onset_test_node",log_level=rospy.DEBUG)
```

```python
topic_rate = rospy.Rate(100)
buffer = np.array([], dtype=float)
bag = rosbag.Bag(bag_file)
for topic, msg, t in bag.read_messages(topics=['/audio_node/audio_stamped']):
    trunk = np.array(struct.unpack(f"{int(len(msg.audio.data) / 2)}h",bytes(msg.audio.data)),dtype=float)
    buffer = np.concatenate([buffer,trunk])
bag.close()
```

```python
print(f"buffer size:{buffer.shape}")
print(f"audio length:{buffer.shape[0]/sr}")
```

```python
plt.figure(figsize=(14, 5))
librosa.display.waveshow(buffer, sr=sr)
# plt.vlines(onset_times, -0.8, 0.79, color='r', alpha=0.8) 
```

```python
def get_buffer(buffer, start_t,end_t):
    return buffer[start_t*sr:end_t*sr]

data = get_buffer(buffer,5,17)
plt.figure(figsize=(14, 5))
librosa.display.waveshow(data, sr=sr)
# plt.vlines(onset_times, -0.8, 0.79, color='r', alpha=0.8) 
```

```python
def constant_q_transform(buffer):
    cqt = np.abs(
        librosa.cqt(
            y=buffer,
            sr=sr,
            hop_length=hop_length,
            fmin=fmin,
            n_bins=semitones,
        )
    )
    return cqt
```

```python
cqt = constant_q_transform(data)
```

```python
print(f"cqt shape:{cqt.shape}")
```

```python
librosa.display.specshow(np.log(cqt+1e-8), sr=sr, x_axis='time', y_axis='log')
```

```python
S = librosa.stft(data)
logS = librosa.amplitude_to_db(S)
onset_frames = librosa.onset.onset_detect(data, sr=sr, hop_length=hop_length, wait=1, pre_avg=1, post_avg=1, pre_max=1, post_max=1)
print(onset_frames.shape)
librosa.display.specshow(logS, sr=sr, x_axis='time', y_axis='log')
onset_times = librosa.frames_to_time(onset_frames, sr=sr, hop_length=hop_length)
plt.vlines(onset_times, 0, 10000, color='white')
```

```python
onset_envelope = librosa.onset.onset_strength(data, sr=sr, hop_length=hop_length)
onset_envelope.shape
```

```python
N = len(data)
T = N/float(sr)
t = np.linspace(0, T, len(onset_envelope))
plt.figure(figsize=(14, 5))
plt.plot(t, onset_envelope)
plt.xlabel('Time (sec)')
plt.xlim(xmin=0)
plt.ylim(0)
```

```python
# Adjust the peak_pick parameters, pre_max, post_max, pre_avg, post_avg, delta, and wait. How do the detected peaks change?
"""
Parameters
:
xnp.ndarray [shape=(n,)]
input signal to peak picks from

pre_maxint >= 0 [scalar]
number of samples before n over which max is computed

post_maxint >= 1 [scalar]
number of samples after n over which max is computed

pre_avgint >= 0 [scalar]
number of samples before n over which mean is computed

post_avgint >= 1 [scalar]
number of samples after n over which mean is computed

deltafloat >= 0 [scalar]
threshold offset for mean

waitint >= 0 [scalar]
number of samples to wait after picking a peak
"""
```

```python
onset_frames = librosa.util.peak_pick(
    onset_envelope, 
    pre_max=7, 
    post_max=7, 
    pre_avg=7,
    post_avg=7, 
    delta=0.5, 
    wait=5)
plt.figure(figsize=(14, 5))
plt.plot(t, onset_envelope)
plt.grid(False)
plt.vlines(t[onset_frames], 0, onset_envelope.max(), color='r', alpha=0.7)
plt.xlabel('Time (sec)')
plt.xlim(0, T)
plt.ylim(0)
```
